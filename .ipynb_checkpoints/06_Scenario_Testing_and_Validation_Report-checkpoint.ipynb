```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score, precision_score, recall_score
import warnings
warnings.filterwarnings('ignore')

# Load data and model
print("Loading data and model...")
X_test = pd.read_csv('X_test.csv')
y_test = pd.read_csv('y_test.csv').squeeze()
X_train = pd.read_csv('X_train.csv')
y_train = pd.read_csv('y_train.csv').squeeze()

model = XGBClassifier(
    n_estimators=300,
    max_depth=6,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    scale_pos_weight=27,
    random_state=42,
    eval_metric='aucpr',
    n_jobs=-1
)
model.fit(X_train, y_train, verbose=False)

y_pred_proba = model.predict_proba(X_test)[:, 1]
optimal_threshold = 0.83

print("Ready for scenario testing.")
print(f"Baseline ROC-AUC: {roc_auc_score(y_test, y_pred_proba):.4f}")
```

```python
# Scenario Testing
# SR 11-7: Model must be tested under stress conditions
# to assess robustness and identify limitations

scenarios = {}

# Scenario 1: High-value transactions only (>$500)
mask_hv = X_test['TransactionAmt'] > 500
if y_test[mask_hv].sum() > 30:
    auc = roc_auc_score(y_test[mask_hv], y_pred_proba[mask_hv])
    scenarios['High Value (>$500)'] = {
        'N': mask_hv.sum(),
        'Fraud Cases': y_test[mask_hv].sum(),
        'ROC-AUC': round(auc, 4)
    }

# Scenario 2: Low-value transactions (<$50)
mask_lv = X_test['TransactionAmt'] < 50
if y_test[mask_lv].sum() > 30:
    auc = roc_auc_score(y_test[mask_lv], y_pred_proba[mask_lv])
    scenarios['Low Value (<$50)'] = {
        'N': mask_lv.sum(),
        'Fraud Cases': y_test[mask_lv].sum(),
        'ROC-AUC': round(auc, 4)
    }

# Scenario 3: Transactions without identity data
mask_no_id = X_test['has_identity'] == 0
if y_test[mask_no_id].sum() > 30:
    auc = roc_auc_score(y_test[mask_no_id], y_pred_proba[mask_no_id])
    scenarios['No Identity Data'] = {
        'N': mask_no_id.sum(),
        'Fraud Cases': y_test[mask_no_id].sum(),
        'ROC-AUC': round(auc, 4)
    }

# Scenario 4: Transactions with identity data
mask_id = X_test['has_identity'] == 1
if y_test[mask_id].sum() > 30:
    auc = roc_auc_score(y_test[mask_id], y_pred_proba[mask_id])
    scenarios['With Identity Data'] = {
        'N': mask_id.sum(),
        'Fraud Cases': y_test[mask_id].sum(),
        'ROC-AUC': round(auc, 4)
    }

# Scenario 5: Night transactions (00:00 - 09:00)
mask_night = X_test['is_night'] == 1
if y_test[mask_night].sum() > 30:
    auc = roc_auc_score(y_test[mask_night], y_pred_proba[mask_night])
    scenarios['Night Transactions'] = {
        'N': mask_night.sum(),
        'Fraud Cases': y_test[mask_night].sum(),
        'ROC-AUC': round(auc, 4)
    }

# Scenario 6: High-risk email domain
mask_email = X_test['is_highrisk_email'] == 1
if y_test[mask_email].sum() > 30:
    auc = roc_auc_score(y_test[mask_email], y_pred_proba[mask_email])
    scenarios['High-Risk Email'] = {
        'N': mask_email.sum(),
        'Fraud Cases': y_test[mask_email].sum(),
        'ROC-AUC': round(auc, 4)
    }

# Results
df_scenarios = pd.DataFrame(scenarios).T.reset_index()
df_scenarios.columns = ['Scenario', 'N', 'Fraud Cases', 'ROC-AUC']
df_scenarios['Fraud Rate (%)'] = (df_scenarios['Fraud Cases'] / df_scenarios['N'] * 100).round(2)
df_scenarios['vs Baseline'] = (df_scenarios['ROC-AUC'] - 0.9356).round(4)

print("=== SCENARIO TESTING RESULTS ===")
print(f"Baseline ROC-AUC: 0.9356\n")
print(df_scenarios[['Scenario', 'N', 'Fraud Rate (%)', 'ROC-AUC', 'vs Baseline']].to_string(index=False))
```

```python
# Scenario Testing Visualization
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: ROC-AUC by scenario
colors = ['crimson' if x < 0.9356 else 'steelblue' 
          for x in df_scenarios['ROC-AUC']]
bars = axes[0].barh(df_scenarios['Scenario'], 
                    df_scenarios['ROC-AUC'], color=colors)
axes[0].axvline(x=0.9356, color='gray', linestyle='--', 
                label='Baseline (0.9356)')
axes[0].set_title('ROC-AUC by Scenario\n(Red = below baseline)')
axes[0].set_xlabel('ROC-AUC')
axes[0].set_xlim(0.85, 0.98)
axes[0].legend()
for bar, val in zip(bars, df_scenarios['ROC-AUC']):
    axes[0].text(val + 0.001, bar.get_y() + bar.get_height()/2,
                f'{val:.4f}', va='center', fontsize=9)

# Plot 2: Delta vs baseline
colors2 = ['crimson' if x < 0 else 'steelblue' 
           for x in df_scenarios['vs Baseline']]
bars2 = axes[1].barh(df_scenarios['Scenario'], 
                     df_scenarios['vs Baseline'], color=colors2)
axes[1].axvline(x=0, color='gray', linestyle='--')
axes[1].set_title('Performance Delta vs Baseline\n(Red = degradation)')
axes[1].set_xlabel('ROC-AUC Delta')
for bar, val in zip(bars2, df_scenarios['vs Baseline']):
    offset = 0.001 if val >= 0 else -0.005
    axes[1].text(val + offset, bar.get_y() + bar.get_height()/2,
                f'{val:+.4f}', va='center', fontsize=9)

plt.suptitle('SR 11-7 Scenario Testing — Model Robustness Analysis',
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('12_scenario_testing.png', dpi=150, bbox_inches='tight')
plt.show()
print("Saved: 12_scenario_testing.png")
```
